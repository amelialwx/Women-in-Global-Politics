**Women in Parliament**

```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

Preprocessing for visualization #1: Average Percentage of Women in Parliament by Continent Over Time

```python
countries_by_share_of_women_in_parliament = pd.read_csv('/content/countries-by-share-of-women-in-parliament.csv')
countries_by_share_of_women_in_parliament.head()
```

```python
countries_by_share_of_women_in_parliament_grouped_by_year = countries_by_share_of_women_in_parliament.groupby('Year').aggregate('sum')
countries_by_share_of_women_in_parliament_grouped_by_year.head()
```

```python
countries_by_share_of_women_in_parliament_grouped_by_year.to_csv('/content/countries-by-share-of-women-in-parliament-grouped-by-year.csv')
```

Preprocessing for visualization #2: Number of Countries with Share of Women in Parliament Over Time

```python
distribution_of_women_share_in_parliament = pd.read_csv('/content/distribution-of-women-share-in-parliament.csv')
distribution_of_women_share_in_parliament.head()
```

```python
distribution_of_women_share_in_parliament['Continent'].str.strip().dropna().unique().tolist()
```

```python
distribution_of_women_share_in_parliament = distribution_of_women_share_in_parliament[distribution_of_women_share_in_parliament['Entity'].isin(distribution_of_women_share_in_parliament['Continent'].str.strip().dropna().unique().tolist())]
distribution_of_women_share_in_parliament.head()
```

```python
distribution_of_women_share_in_parliament = pd.pivot_table(distribution_of_women_share_in_parliament, values=['Lower chamber female legislators (aggregate: average)'], index=['Year'], columns=['Entity'])
```

```python
distribution_of_women_share_in_parliament.to_csv('/content/distribution-of-women-share-in-parliament-continent-pivot.csv', index=False)
```
